```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
institutionDataFilepath = os.path.join(os.getcwd(), 'datasets', 'CollegeScorecardMostRecentInstitution06102024', 'Most-Recent-Cohorts-Institution.csv')

institutionDataDF = pd.read_csv(institutionDataFilepath, low_memory=False)
institutionDataDF.columns.to_list()
```

```python
# find necessary columns
keptColumns = ["INSTNM", "CITY", "ACCREDAGENCY", "HCM2", "MAIN", "PREDDEG", "HIGHDEG", "CONTROL", "ST_FIPS", "REGION", "CCBASIC", "CCUGPROF", "ADM_RATE"]
renamedColumns = {
	"INSTNM": "institute_name", "CITY": "city", "ST_FIPS": "state_num", "REGION": "region",
	"ACCREDAGENCY": "accredited", "HCM2": "financially_monitored", 
	"CONTROL": "organization", "MAIN": "is_main_campus", 
	"PREDDEG": "predominant_deg_awarded", "HIGHDEG": "highest_deg_awarded", "ADM_RATE": "admission_rate",
	"CCBASIC": "college_type", "CCSIZSET": "college_length_and_setting"
}

# ACCREDAGENCY | accredited - will be changed to 1 for credited, 0 for non-credited
# HCM2 | financially_monitored - the more severe HCM2 requires colleges to pay out money to students before being reimbursed by Federal Student Aid
# CONTROL | organization - public 1, private non-profit 2, private for-profit 3
# ST_FIPS | state_num - state as a number (see data dictonary)
# REGION | region - the region the institute resides in (see data dictonary)
# 

cleanInstitutionDataDF = institutionDataDF[keptColumns].rename(columns=renamedColumns)

cleanInstitutionDataDF.head()
```
